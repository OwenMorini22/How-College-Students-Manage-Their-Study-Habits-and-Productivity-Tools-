```python
whoDF = pd.read_csv('Study Management - Form Responses 1.csv')

whoDF.columns = whoDF.columns.str.strip()
total_responses = len(whoDF)

print(f"Total Responses: {total_responses}")
```

WHO: Percentages of each who question.

Q1.

```python
class_year_counts = whoDF['1. What is your class year?'].value_counts()
print("\nCounts:")
print(class_year_counts)

```

```python
class_year_percentages = (class_year_counts / len(whoDF) * 100).round(2)
print("\nPercentages:")
print(class_year_percentages)
```

Q2.

```python
major_counts = whoDF['2. What is your major?'].value_counts()
print("\nMajor Counts:")
print(major_counts)
```

```python
major_percentages = (major_counts / len(whoDF) * 100).round(2)
print("\nMajor Percentages:")
print(major_percentages)
```

Q3. 

```python
gender_counts = whoDF['3. What is your gender?'].value_counts() 
print("\nGender Counts:")
print(gender_counts)    
```

```python
gender_percentages = (gender_counts / len(whoDF) * 100).round(2)
print("\nGender Percentages:")  
print(gender_percentages)
```

WHAT: What percentage of users checked each reply from the what question? 

Q4. What primary method do you use to organize your studying?

```python
organizedMeth = whoDF.groupby('4. What primary method do you use to organize your studying?')

organizedMeth.size()

```

```python
organizedMeth_counts = whoDF['4. What primary method do you use to organize your studying?'].value_counts() 

organizedMeth_percentages = (organizedMeth_counts / len(whoDF) * 100).round(2)
print("\nOrganized Study Method Percentages:")  
print(organizedMeth_percentages)


plt.pie(x=organizedMeth_counts, 
        labels=organizedMeth_counts.index,
         autopct='%1.1f%%',
         colors=['blue', 'green', 'coral', 'orange', 'pink', 'red'])
plt.title('Primary Method Used to Organize Studying')   
plt.show()
```

WHEN: Percentage of each option

Q5. When do you typically do most of your studying?  

```python
whenStudy_counts = whoDF.groupby('5. When do you typically do most of your studying?')

whenStudy_percentages = (whenStudy_counts.size() / len(whoDF) * 100).round(2)
print("\nWhen do you study most Percentages:")  
print(whenStudy_percentages)

plt.pie(whenStudy_percentages, 
        labels=whenStudy_percentages.index, 
        autopct='%1.1f%%',
        colors=['lightyellow', 'lightblue', 'lightpink', 'lightgreen'])


plt.title('When Do You Typically Do Most of Your Studying?')
plt.show()
#percentages are rounded
```

Q8. Why do you study at the time of day you selected?

Q9. Why do you study in the location you selected?

```python
whyStudyLoc = whoDF.groupby('9. Why do you study in the location you selected?')
whyStudyLoc.size()
whyStudyLoc_percentages = (whyStudyLoc.size() / len(whoDF) * 100).round(2)
print("\nWhy do you study in the location you selected Percentages:")
print(whyStudyLoc_percentages)

plt.pie(whyStudyLoc_percentages,
        labels=whyStudyLoc_percentages.index,
        autopct='%1.1f%%',
        colors=["#F5F102", "#F90101", "#002FFF", "#00941E"])

plt.title('Why Do You Study in the Location You Selected?')
plt.show()
```

```python
avgStudyHours_counts = whoDF['10. On average, how many hours per week do you study?'].value_counts().sort_index()
print("\nAverage Study Hours per Week:")
print(avgStudyHours_counts.describe())



```

```python
plt.Figure(figsize=(10,6))
plt.bar(avgStudyHours_counts.index, avgStudyHours_counts.values, color=['#FFA500', '#FF4500', '#32CD32', '#1E90FF', '#8A2BE2'])
plt.xlabel('Average Study Hours per Week')
plt.ylabel('Number of Students')
plt.title('Average Study Hours per Week Distribution')
plt.tight_layout()
plt.show()
```
